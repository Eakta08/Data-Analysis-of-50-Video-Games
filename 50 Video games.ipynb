## Data Analysis

#### (i) How many Games are there in each genre?

```python
ax=sns.countplot(y=data.Genre,data=data)
for i in ax.containers:
    ax.bar_label(i,fmt='%g')
plt.xlabel("No. of video games")
plt.ylabel("Genre")
plt.show()
```

> _The above graph indicates the no. of video games in each genre. Also indicating in Shooter genre highest no. of video games are present i.e 10 and lowest no. of video games are present in Fighting genre i.e. 1 ._

```python
df=data.Genre.value_counts()
df
```

```python
plt.pie(df,labels=df.index,autopct='%.0f%%',startangle=65,shadow=True
       ,explode=[0.035 for i in range(len(df.index))])
plt.show()
```

> _The above pie chart shows the contribution of each genre (percentage wise). Shooter genre has the highest contribution i.e. 20% and Fighting genre has the lowest contribution i.e. 2%._

> _The above graph indicates the no. of video games published by each publisher. Also indicating in Nintendo highest no. of video games are present i.e 32 and lowest no. of video games are present in Microsoft Game Studios & Sony Computer Entertainment i.e. 2 ._

> * _NA_Sales: Four outliers are present. Minimum is about 2, maximum is over 41, and median is almost 8. About 11 is the upper quantile, and 6 is the lower quantile._
> * _JP_Sales: Only one outlier is present. Minimum is about 0, maximum is over 10, and median is almost 3. About 4 is the upper quantile, and 0 is the lower quantile._
> * _EU_Sales: Two outliers are present. Minimum is about 0, maximum is over 29, and median is almost 4. About 7 is the upper quantile, and 3 is the lower quantile_
> * _Other_Sales: Three outliers are present. Minimum is about 0, maximum is over 11, and median is almost 1. About 2 is the upper quantile, and 0 is the lower quantile._
> * _Global_Sales: Two outliers are present. Minimum is about 11, maximum is over 83, and median is almost 16. About 23 is the upper quantile, and 13 is the lower quantile._

#### (viii) Performance of each Publisher year wise
* NA_Sales
* EU_Sales
* JP_Sales
* Other_Sales
* Global_Sales

```python
Publisher=data.Publisher.unique()
for i in range(len(Publisher)):
    df=data[data.Publisher==Publisher[i]].sort_values(by='Year')
    sns.lineplot(x='Year',y='NA_Sales',data=df,marker='o')
    sns.lineplot(x='Year',y='EU_Sales',data=df,marker='o')
    sns.lineplot(x='Year',y='JP_Sales',data=df,marker='o')
    sns.lineplot(x='Year',y='Other_Sales',data=df,marker='o')
    sns.lineplot(x='Year',y='Global_Sales',data=df,marker='o')
    plt.legend(["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"],loc='upper left')
    plt.ylabel("Sales")
    plt.title(Publisher[i])
    plt.show()
```

> _The above graphs illustrates the trends of sales done by each publisher per year. The number of markers in each graph denotes the no. of video games published by those publishers._

> _The above graphs illustrates the trends of sales done by each platform per year. The number of markers in each graph denotes the no. of video games launched on that particular platform._
